```python
# Importando bibliotecas.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import urllib
```

```python
# Extração: Coletar dados de uma fonte.

# Lista de URLs dos arquivos CSV

urls = [
    "/content/drive/MyDrive/Colab Notebooks/Visualização de dados/Desafio - Visualização de Dados Streamlit/sales_data.csv"
]

# Create a list to store DataFrames
dfs = []

# Iterate through the URLs and read each CSV into a DataFrame
for url in urls:
    df = pd.read_csv(url)  # Assuming your data is in CSV format
    dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
df_final = pd.concat(dfs, ignore_index=True)

# Visualização inicial dos dados com os novos nomes de coluna
print("Visualização inicial dos dados:")
df_final.head()
```
